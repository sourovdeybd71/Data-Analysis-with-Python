```python
#Import required libraries
import numpy as np
import pandas as pd
```

```python
#reading excel file
df= pd.read_excel("test_cricket.xlsx", sheet_name='runs')

display(df.head())
```

#### rename columns name

```python
print(df.columns)
```

```python
df=df.rename(columns={'Mat':'Match', 
                        'Inns':'Innings',
                        'NO': 'NotOut',
                        'HS': 'Highest_score',
                        'Ave': 'Average',
                        100: 'Centuries',
                        50: 'Half_centuries',
                        0: 'Ducks'})
display(df.head())
```

### Creating a new DataFrame

```python
df_A = pd.DataFrame({"A":[1,2,3],
                    "B":[4,5,6]})
display(df_A)
```

```python
df_B= pd.DataFrame()
df_B["A"]=[7,8,9]
df_B["B"]=[0,11,12]
display(df_B)
```

### Concatenate two dataframes

```python
#column-wise concatenation
df_C = pd.concat([df_A, df_B], axis=1)
display(df_C)
```

```python
#row-wise concatenation
df_C = pd.concat([df_A, df_B], axis=0)
display(df_C)
```

#### How to split a column and create two new column

```python
df_player = df['Player'].str.split("(",expand=True)
display(df_player.head())
```

```python
df = pd.concat([df, df_player], axis=1)
display(df.head())
```

```python
#removing Player colmn
df = df.drop("Player", axis=1)
df = df.rename(columns={0: 'Player',
                        1: 'Country'})
display(df.head())
```

```python
#replace/remove a value from a pandas colmn
df["Country"] = df['Country'].str.replace(")", "")
display(df.head())
```

```python
print(df.columns)
new_col_sequence = ['Player', 'Country', 'Span', 'Match', 'Innings', 'NotOut', 'Runs', 'Highest_score',
       'Average', 'Centuries', 'Half_centuries', 'Ducks']
```
