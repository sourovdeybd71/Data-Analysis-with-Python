# Data Joining using PANDAS

### Objectives: <br>
> Inner join, Outer Join, Left join, Right Join

```python
import pandas as pd
```

### Creating two random dataframes

```python
df_A = pd.DataFrame()
df_A['Name'] = ['Shuvo', 'Tanmoy', 'Avishek']
df_A['Address']= ['New York', 'Brisbane', 'Bali']
df_A["University"] = ['LU', 'LU', 'CUET']

display(df_A)
```

```python
df_B = pd.DataFrame()
df_B['Name'] = ['Shotabdy', 'Sourov', 'Tanmoy']
df_B['Address'] = ['Toronto', 'Las Vegas', 'Brisbane']
df_B['Department'] = ['CE', "CSE", "CSE"]

display(df_B)
```

## Inner JOIN<br>
>The INNER JOIN keyword selects records that have matching values in both tables. 

```python
df_inner_ex1 =df_A.merge(df_B, how="inner", on='Name')
df_inner_ex2 =df_A.merge(df_B, how="inner", on='Address')
df_inner_ex3 =df_A.merge(df_B, how="inner")

display(df_inner_ex1)
display(df_inner_ex2)
display(df_inner_ex3)
```

## Outer JOIN<br>
>OUTER JOIN returns unmatched rows from both tables.

```python
df_outer_e1 = df_A.merge(df_B, how='outer')
df_outer_e2 = df_A.merge(df_B, how='outer', on='Name')
df_outer_e3 = df_B.merge(df_A, how='outer', on='Address')

display(df_outer_e1)
display(df_outer_e2)
display(df_outer_e3)
```

## Left JOIN<br>
> LEFT JOIN returns only unmatched rows from the left table

```python
df_left_e1 = df_A.merge(df_B, how='left')
df_left_e2 = df_A.merge(df_B, how='left', on='Name')
df_left_e3 = df_A.merge(df_B, how='left', on='Address')

display(df_left_e1)
display(df_left_e2)
display(df_left_e3)
```

```python
df_left_e4 = df_B.merge(df_A, how='left', on='Name')

display(df_left_e4)
```

## Right JOIN<br>
>RIGHT JOIN returns only unmatched rows from the right table

```python
df_right = df_B.merge(df_A, how='right', on='Name')

display(df_right)

# df_right = df_A.merge(df_B, how='right', left_on='name', right_on='Name')
# display(df_right)
```

### DAta Joinig on FRIENDS dataset

```python
friends_1 = pd.read_excel("friends.xlsx", sheet_name="birthday")
friends_2 = pd.read_excel("friends.xlsx", sheet_name = 'gender')

display(friends_1)
display(friends_2)
```

#### INNER

```python
inner_join = friends_1.merge(friends_2, 
                             how='inner', 
                             left_on="character", 
                             right_on= "character_name")
display(inner_join)
```

#### OUTER

```python
friends_outer = friends_1.merge(friends_2,
                           how='outer', 
                           left_on='character', 
                           right_on='character_name')

display(friends_outer)
```

```python
friends_left_1 = friends_1.merge(friends_2, 
                           how='left', 
                           left_on='character', 
                           right_on='character_name')

display(friends_left_1)
```

```python
friends_left_2 = friends_2.merge(friends_1, 
                           how='left', 
                           right_on='character', 
                           left_on='character_name')

display(friends_left_2)
```

```python
friends_right_1 = friends_1.merge(friends_2, 
                           how='right', 
                           left_on='character', 
                           right_on='character_name')

friends_right_2 = friends_2.merge(friends_1, 
                           how='right', 
                           right_on='character', 
                           left_on='character_name')

display(friends_right_1)
display(friends_right_2)
```
